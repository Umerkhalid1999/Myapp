**Project Flow**

```python
from sklearn.model_selection import train_test_split # for splitting the data into training and testing sets
from sklearn.linear_model import LogisticRegression  # for creating a logistic regression model
import pandas as pd # for structuring raw dataset into tabular form 
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
# import few classification algorithms for training the model and evaluating their performance 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("diabetes.csv") # 1. Load Dataset 
```

```python
df.head() # for checking the first 5 rows of the dataset
```

```python
# 3. Splitting the data into training and testing sets 
X = df.drop(columns= "Outcome", axis=1) # dropping the "Outcome" column from the dataset and storing the remaining columns in variable X
Y = df['Outcome'] # storing the "Outcome" column in variable Y
```

```python
X.head()
```

```python
Y.head()
```

```python
X_scaled = StandardScaler().fit_transform(X) # Normalization of the data
```

```python
# 4. Splitting the data into training and testing sets using train_test_split 
X_scaled, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.2, random_state=42)
```

```python
models = {
            "LogisticRegression": LogisticRegression(max_iter=5000),
            "DecisionTree": DecisionTreeClassifier(),
            "RandomForest": RandomForestClassifier(),
            "SVC": SVC(),
            "KNeighbors": KNeighborsClassifier()
        }
```

```python
 
# 7. Evaluating the model score    

for model_name, model in models.items(): 
      model.fit(X_scaled, y_train)                               # training the model using the fit method. 

      Training_score = model.score(X_train, y_train)            # Checking score on training Data 
      Testing_score = model.score(X_test, y_test)               # Checking score on Testing Data   
       
      average = (Training_score + Testing_score) / 2    

      print(f"Model: {model_name}")  
      print(f"The score on Training Data:  {Training_score * 100:.2f} % ") 
      print(f"The score on Testing Data:  {Testing_score * 100:.2f} % ")   
      print(f"The average score is: {average * 100:.2f} % ")
      print("\n")  
```

```python
# 5. Training the best model from above RandomForestClassifier 
# Model: RandomForest
# The score on Training Data:  100.00 % 
# The score on Testing Data:  74.03 % 
#The average score is: 87.01 %  

model = RandomForestClassifier() # Algorithm: LogisticRegression()
model.fit(X_scaled, y_train) # training the logistic regression model using the fit method, which takes the training data (X_train) and corresponding labels (y_train) as input
```

```python
model.score(X_scaled, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# [1,	89,	66,	23,	94,	28.1,	0.167,	21] 
model.predict([[1,	89,	66,	23,	94,	28.1,	0.167,	21]] 
)
```

```python
# 6. Performing Predictions 
# he model will output the predicted class label for this input data point. 

if model.predict([[1,89,66,	23,	94,	28.1,	0.167,	21]]) == 0: 
    print("The person is not Diabetic") 
elif model.predict([[1,89,66,	23,	94,	28.1,	0.167,	21]]) == 1: 
    print("The person is Diabetic")
```
